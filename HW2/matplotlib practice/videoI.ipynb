```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ages_x = np.linspace(25, 35, 11, dtype=int)
```

```python
dev_y = np.array([38496, 42000, 46752, 49320, 53200, 
        56000, 62316, 64928, 67317, 68748, 73752])
py_dev_y = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
js_dev_y = [37810, 43515, 46823, 49293, 53437,
            56373, 62375, 66674, 68745, 68746, 74583]
```

```python
plt.xkcd()
plt.plot(ages_x, dev_y, color='#444444', linestyle='--', label='All Devs', linewidth=3)
plt.plot(ages_x, py_dev_y, color='#5a7b9a', label='Python', linewidth=1)
plt.plot(ages_x, js_dev_y, color='#adad3b', label='JavaScript')
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.title('Median Salary (USD) by Age')
plt.legend()
plt.tight_layout()
plt.savefig('ex1.png')
plt.show()
```
